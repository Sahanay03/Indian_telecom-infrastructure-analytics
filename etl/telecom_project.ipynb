```python
infra = infra.fillna(0)
```

```python
#Standardizing column names
```

```python
telecom.columns = telecom.columns.str.lower().str.replace(" ", "_")
density.columns = density.columns.str.lower().str.replace(" ", "_")
```

```python
total.columns = total.columns.str.lower().str.replace(" ", "_")
wireline.columns = wireline.columns.str.lower().str.replace(" ", "_")
infra.columns = infra.columns.str.lower().str.replace(" ", "_")
```

```python
# Renaming the columns
```

```python
density = density.rename(columns={
    'tele-density_service_area-wise_(uom:number),_scaling_factor:1000000': 'tele_density'
})
```

```python
wireline = wireline.rename(columns={
    'number_of_wireline_telephones__(uom:number),_scaling_factor:1000000': 'wireline_phones'
})
```

```python
total = total.rename(columns={
    'total_telephone_connections__(uom:number),_scaling_factor:1000000': 'total_connections'
})
```

```python
telecom = telecom.rename(columns={
    'wireless_subscribers_(uom:number),_scaling_factor:1': 'wireless_subscribers'
})
telecom = telecom.rename(columns={
    'wireline_subscribers_(uom:number),_scaling_factor:1': 'wireline_subscribers'
})
```

```python
# Standardize State Names
```

```python
density = density.rename(columns={"service_area":"state"})
wireline = wireline.rename(columns={"service_area":"state"})
total = total.rename(columns={"name_of_the_service_area_having_telephone_connections":"state"})
infra = infra.rename(columns={"state_name":"state"})
```

```python
# Convert year & month
```

```python
telecom.columns = telecom.columns.str.strip()
telecom["year"] = telecom["year"].str.extract(r'(\d{4})').astype(int)
telecom["month"] = telecom["month"].str.extract(r'([A-Za-z]+)')
```

```python
total.columns = total.columns.str.strip()
total["year"] = total["year"].str.extract(r'(\d{4})').astype(int)
total["month"] = total["month"].str.extract(r'([A-Za-z]+)')
```

```python
density.columns = density.columns.str.strip()
density["year"] = density["year"].str.extract(r'(\d{4})').astype(int)
density["month"] = density["month"].str.extract(r'([A-Za-z]+)')
```

```python
wireline.columns = wireline.columns.str.strip()
wireline["year"] = wireline["year"].str.extract(r'(\d{4})').astype(int)
wireline["month"] = wireline["month"].str.extract(r'([A-Za-z]+)')
```

```python
infra["year"] = infra["year"].str.extract(r'(\d{4})').astype(int)
```

```python
# Preparation Telecom Operator Data
```

```python
telecom_state = telecom.groupby(
    ["state","year","month"],
    as_index=False
).agg({
    "wireless_subscribers":"sum",
    "wireline_subscribers":"sum",
    "proportion_of_vlr_on_a_peak_day_(uom:%(percentage)),_scaling_factor:1":"mean"
})
```

```python
telecom_fact = telecom_state.merge(
    density[["state","year","month","tele_density"]],
    on=["state","year","month"],
    how="left"
)

telecom_fact = telecom_fact.merge(
    wireline[["state","year","month","wireline_phones"]],
    on=["state","year","month"],
    how="left"
)

telecom_fact = telecom_fact.merge(
    total[["state","year","month","total_connections"]],
    on=["state","year","month"],
    how="left"
)
```

```python
# Preparation Infrastructure table
```

```python
infra_fact = infra[[
    "state","year",
    "number_of_service_ready_gram_panchayats_including_block_headquarters_phase_i___(uom:number),_scaling_factor:1",
    "number_of_service_ready_gram_panchayats_(gps)_including_block_headquarters_(bhq)_of_phase_-_ii_under_bharat_broadband_network_limited_(bbnl)_(uom:number),_scaling_factor:1"
]]

infra_fact = infra_fact.rename(columns={
    "number_of_service_ready_gram_panchayats_including_block_headquarters_phase_i___(uom:number),_scaling_factor:1":"gp_ready_phase1",
    "number_of_service_ready_gram_panchayats_(gps)_including_block_headquarters_(bhq)_of_phase_-_ii_under_bharat_broadband_network_limited_(bbnl)_(uom:number),_scaling_factor:1":"gp_ready_phase2"
})

infra_fact["gp_ready_total"] = infra_fact["gp_ready_phase1"] + infra_fact["gp_ready_phase2"]
```

```python
final_fact = telecom_fact.merge(
    infra_fact[["state","year","gp_ready_total"]],
    on=["state","year"],
    how="left"
)
```

```python
# Dimension tables
```

```python
dim_time = final_fact[["year","month"]].drop_duplicates().copy()

month_map = {
    "January":1,"February":2,"March":3,"April":4,"May":5,"June":6,
    "July":7,"August":8,"September":9,"October":10,"November":11,"December":12
}

dim_time["month_num"] = dim_time["month"].map(month_map)
dim_time["quarter"] = ((dim_time["month_num"]-1)//3)+1
dim_time["time_id"] = range(1, len(dim_time)+1)
```

```python
dim_location = final_fact[["state"]].drop_duplicates().copy()
dim_location["country"] = "India"
dim_location["location_id"] = range(1, len(dim_location)+1)
```

```python
dim_service = pd.DataFrame({
    "service_id":[1,2,3,4,5],
    "service_name":[
        "Wireless Subscribers",
        "Wireline Subscribers",
        "Tele Density",
        "Total Connections",
        "Gram Panchayats Connected"
    ]
})
```

```python
# Convert wide fact to long fact table
```

```python
fact_long = final_fact.melt(
    id_vars=["state","year","month"],
    value_vars=[
        "wireless_subscribers",
        "wireline_subscribers",
        "tele_density",
        "total_connections",
        "gp_ready_total"
    ],
    var_name="service",
    value_name="value"
)
```

```python
service_map = {
    "wireless_subscribers":1,
    "wireline_subscribers":2,
    "tele_density":3,
    "total_connections":4,
    "gp_ready_total":5
}
```

```python
fact_long["service_id"] = fact_long["service"].map(service_map)
fact_long = fact_long.merge(dim_time, on=["year","month"], how="left")
fact_long = fact_long.merge(dim_location, on="state", how="left")
fact_final = fact_long[["time_id","location_id","service_id","value"]]
```

```python
# Exporting the files to a folder in the system
```

```python
folder_name = "Data/processed/fact.csv"

fact_final.to_csv(os.path.join(folder_name, "fact_telecom_infra.csv"), index=False)
dim_time.to_csv(os.path.join(folder_name, "dim_time.csv"), index=False)
dim_location.to_csv(os.path.join(folder_name, "dim_location.csv"), index=False)
dim_service.to_csv(os.path.join(folder_name, "dim_service.csv"), index=False)

print("All files have been saved to the folder.")
```
